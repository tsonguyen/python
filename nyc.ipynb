```python
# import data from http://www.transtats.bts.gov/DL_SelectFields.asp?Table_ID=236&DB_Short_Name=On-Time
import pandas as pd
fdf_nyc = pd.DataFrame()
list_ = []
for i in range(12):
    fdf=pd.read_csv(str(i+1)+'.csv')
    list_.append(fdf)
fdf_nyc = pd.concat(list_)
```

```python
wk=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
mn=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
for i in range(7):fdf_nyc.loc[fdf_nyc['DAY_OF_WEEK']==(i+1),'DAY_OF_WEEK']=wk[i]
for i in range(12):fdf_nyc.loc[fdf_nyc['MONTH']==(i+1),'MONTH']=mn[i]
fdf_frnyc=fdf_nyc.loc[fdf_nyc['ORIGIN_CITY_NAME']=='New York, NY']
fdf_tonyc=fdf_nyc.loc[fdf_nyc['DEST_CITY_NAME']=='New York, NY']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
# for all flights in NYC
plt.figure(figsize=(20, 10))
plt.legend()
sns.barplot(x='DAY_OF_WEEK',y='DEP_DELAY',hue='MONTH',data=fdf_nyc);
```

```python
# for all flights to NYC
plt.figure(figsize=(20, 10))
plt.legend()
sns.barplot(x='DAY_OF_WEEK',y='DEP_DELAY',hue='MONTH',data=fdf_tonyc);
```

```python
# for all flights from NYC
plt.figure(figsize=(20, 10))
plt.legend()
sns.barplot(x='DAY_OF_WEEK',y='DEP_DELAY',hue='MONTH',data=fdf_frnyc);
```

```python
# all flights in NYC
plt.figure(figsize=(20, 10))
plt.legend()
ax1=sns.barplot(x='DAY_OF_MONTH',y='DEP_DELAY',hue='MONTH',data=fdf_nyc)
ax1.set_xlim(0,9)
plt.figure(figsize=(20, 10))
plt.legend()
ax2=sns.barplot(x='DAY_OF_MONTH',y='DEP_DELAY',hue='MONTH',data=fdf_nyc)
ax2.set_xlim(10,19)
plt.figure(figsize=(20, 10))
plt.legend()
ax3=sns.barplot(x='DAY_OF_MONTH',y='DEP_DELAY',hue='MONTH',data=fdf_nyc)
ax3.set_xlim(20,30)
```
